#### The main goal of our work is to use data of gas consumption and gas prices from the past 4 years (2018 - 2021) to predict consumption and prices for 2022, as if the Ukraine war would have not happened.
#### We will compare this predicted data to the actual data (01 - 07/2022) to show the effect of the war on the consumption and the prices.

### **This notebook contains the data cleaning of the "inflation" datasets.**

```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

### Data from 2018 - 2021

```python
data_original_infl_20xx = pd.read_csv(r"/Volumes/Extreme Pro/WBS Bootcamp Data Science/Final Project/Data/Original/Inflation Europe/Inflation EU_2018-2021_monthly.csv")
data_original_infl_20xx.head()
```

```python
data_original_infl_20xx.info()
```

```python
data_infl_20xx = data_original_infl_20xx.drop(data_original_infl_20xx.iloc[:, 9:],axis = 1)
data_infl_20xx = data_infl_20xx.drop(data_infl_20xx.iloc[:, :7],axis = 1)
data_infl_20xx
```

```python
data_infl_20xx.info()
```

```python
# convert date in datetime-datatype
data_infl_20xx["TIME_PERIOD"] = pd.to_datetime(data_infl_20xx["TIME_PERIOD"])
data_infl_20xx.info()
```

```python
data_infl_20xx = data_infl_20xx.assign(
                            month = data_infl_20xx["TIME_PERIOD"].dt.strftime("%B"),
                            month_no = data_infl_20xx["TIME_PERIOD"].dt.strftime("%m"),
                            year = data_infl_20xx["TIME_PERIOD"].dt.strftime("%Y"))

data_infl_20xx
```

```python
# drop column "TIME_PERIOD"
data_infl_20xx = data_infl_20xx.drop(data_infl_20xx.columns[0],axis = 1)
data_infl_20xx
```

```python
# rename column "OBS_VALUE"
data_infl_20xx = data_infl_20xx.rename(columns={"OBS_VALUE" : "inflation rate"})
data_infl_20xx
```

```python
# change order of columns
data_infl_20xx = data_infl_20xx[["month", "month_no", "year", "inflation rate"]]
data_infl_20xx
```

```python
data_infl_20xx.info()
```

```python
# changing datatype of "month_no" and "year" to integers
data_infl_20xx[["year", "month_no"]] = data_infl_20xx[["year", "month_no"]].apply(pd.to_numeric)
data_infl_20xx.info()
```

```python
# sort rows
data_infl_20xx = data_infl_20xx.sort_values(["month_no", "year"]).reset_index(drop=True)
data_infl_20xx
```

```python
# creating csv-file with final cleaned dataframe
data_infl_20xx.to_csv("Final Data_Inflation EU_2018-2021_monthly.csv")
```
